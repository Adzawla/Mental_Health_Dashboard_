### Let's create a class to wrap up all the steps

```python
class DataCleaner:
    
    def __init__(self, df: pd.DataFrame):
        self.df = df.copy()

    
    def rename_date(self):
        self.df = self.df.rename(columns={"Timestamp": "Date"})
        if pd.api.types.is_datetime64_any_dtype(self.df["Date"]):
            self.df["Date"] = self.df["Date"].dt.strftime("%d/%m/%Y")
        return self
    
    
   
    
    
    def clean_gender(self):
        
        # Step 1: lower + strip
        series_clean = self.df["Gender"].str.lower().str.strip()
    
        # Step 2: mapping dictionary
        mapping = {
        # Male variants
        'm': 'Male', 'male': 'Male', 'man': 'Male', 'male (cis)': 'Male',
        'cis male': 'Male', 'cis man': 'Male', 'mal': 'Male', 'maile': 'Male',
        'make': 'Male', 'msle': 'Male', 'mail': 'Male', 'malr': 'Male',
        'guy (-ish) ^_^': 'Male',

        # Female variants
        'f': 'Female', 'female': 'Female', 'woman': 'Female', 'female ': 'Female',
        'femake': 'Female', 'femail': 'Female', 'cis female': 'Female',
        'cis-female/femme': 'Female', 'female (cis)': 'Female', 
            
            
        # Other / noise
        'nah': 'Other', 'all': 'Other',
        'something kinda male?': 'Other',
        'a little about you': 'Other',
        'p': 'Other', 'neuter': 'Other',
        'non-binary': 'Other', 'enby': 'Other',
        'genderqueer': 'Other', 'fluid': 'Other',
        'androgyne': 'Other', 'agender': 'Other',
        'male leaning androgynous': 'Other',
        'queer/she/they': 'Other', 'queer': 'Other',
        'female (trans)': 'Other',
        'trans-female': 'Other',
        'trans woman': 'Other',
        'ostensibly male': 'Other',
        'ostensibly male': 'Other',
        'unsure what that really means': 'Other',
        "Other/Unknown":'Other',
        }
        self.df["clean_gender"] = series_clean.replace(mapping)
        self.df["clean_gender"] = self.df["clean_gender"].where(self.df["clean_gender"].isin(["Male", "Female"]), "Other")
        return self

 
        
    def classify_company_size(self):
        category_map = {
            "1-5": "vs",
            "6-25": "ME",
            "26-100": "SE",
            "100-500": "ME",
            "500-1000": "BC",
            "More than 1000": "VBC"
        }
        order = [
            "vs",
            "SE",
            "ME",
            "BC",
            "VBC"
        ]
        self.df["company_size"] = (
            self.df["no_employees"]
            .map(category_map)
            .astype(pd.CategoricalDtype(categories=order, ordered=True))
        )
        return self
        
        
    #Inconsistency_removal
    
    
    def inconsistency_removal(self):
        # Replace countries containing "The" with "The Bahamas"
        self.df.loc[self.df["Country"].str.contains("The", case=False, na=False), "Country"] = "The Bahamas"
        
        # Drop rows with missing values in key columns
        self.df = self.df.dropna(subset=["work_interfere", "self_employed"])
        
        ## Drop unecessary columns
        self.df = self.df.drop(columns = [ "comments", "Gender", "state"])
        return self
    
    @staticmethod
    def map_country_to_continent(country_name):
        try:
            country_alpha2 = pc.country_name_to_country_alpha2(country_name)
            continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
            continent_name = pc.convert_continent_code_to_continent_name(continent_code)
            return continent_name
        except:
            return 'Unknown'
    
    def add_continent(self, country_col='Country', continent_col='Continent'):
        
        """
        Adds a new column with continent names based on the Country column.
        """
    
        self.df[continent_col] = self.df[country_col].apply(DataCleaner.map_country_to_continent)
        return self

    
    def get_clean_data(self):
        return self.df
    
```

```python

def plot_sunburst(df, path_cols, color_col=None, title="Sunburst Chart", 
                  width=650, height=450):
    """
    Creates a sunburst chart from a DataFrame.

    Parameters:
    -----------
    df : pd.DataFrame
        The dataset to visualize.
    path_cols : list
        List of columns representing the hierarchy (e.g., ["Continent", "Country"]).
    color_col : str, optional
        Column to color nodes by (default: None).
    title : str, optional
        Title of the chart.
    width : int, optional
        Chart width.
    height : int, optional
        Chart height.

    Returns:
    --------
    fig : plotly.graph_objects.Figure
        The sunburst chart figure.
    """

    fig_state_country = px.sunburst(
        df,
        path=path_cols,
        values=None,
        color=color_col,
        title=title
    )

    fig_state_country.update_layout(
        width=width,
        height=height,
        margin=dict(t=80, l=50, r=150, b=50),
        title=dict(
            x=0.5,  # center
            xanchor="center",
            font=dict(size=24)
        )
    )

    return fig_state_country


# Creating the plot
fig_state_country = plot_sunburst(
    df_clean,
    path_cols=["Continent", "Country"],
    color_col="Continent",
    title="Country by Continent"
)

fig_state_country.show()
```

#### 2- Responses by Country (Top 10)

```python

def plot_top_countries(df, country_col="Country", top_n=10, 
                       width=700, height=450, 
                       title="Top Countries with the Most Responses"):
    """
    Creates a bar chart of the top N countries by response count,
    showing percentage and count labels, with bars styled like cylinders.
    
    Parameters:
    -----------
    df : pd.DataFrame
        The dataset to analyze.
    country_col : str, optional
        Column containing country names.
    top_n : int, optional
        Number of top countries to display (default=10).
    width, height : int
        Dimensions of the chart.
    title : str
        Title of the chart.
    
    Returns:
    --------
    fig : plotly.graph_objects.Figure
        The formatted bar chart.
    """
    
    # Count responses per country and get Top N
    top_countries = (
        df[country_col]
        .value_counts()
        .nlargest(top_n)
        .reset_index()
    )
    top_countries.columns = ['Country', 'Responses']

    # Compute percentages
    total_responses = df.shape[0]
    top_countries['Percent'] = (top_countries['Responses'] / total_responses * 100).round(1)

    # Format labels: "xx% (count)"
    top_countries['Label'] = (
        top_countries['Percent'].astype(str) + '% (' + 
        top_countries['Responses'].astype(str) + ')'
    )

    # Build bar chart
    fig_top_countries = px.bar(
        top_countries,
        x='Country',
        y='Percent',
        text='Label',
        color='Percent',
        color_continuous_scale='Viridis',
        title=title
    )

    # Style bars to look cylindrical
    fig_top_countries.update_traces(
        texttemplate='%{text}',
        textposition='outside',
        marker=dict(
            line=dict(width=1, color="black"),  # outline for 3D feel
            opacity=0.9
        )
    )

    # Layout adjustments
    fig_top_countries.update_layout(
        yaxis_title='Percentage of Responses',
        xaxis_title='Country',
        width=width,
        height=height,
        bargap=0.3,
        margin=dict(t=80, l=50, r=150, b=50),
        title=dict(
            text=title,
            x=0.5,
            xanchor='center',
            font=dict(size=24, color='black', family="Arial, sans-serif")
        )
    )

    return fig_top_countries
  
    
# Creating the plot
fig_top_countries = plot_top_countries(df_clean)
fig_top_countries.show()
```

#### 3- Responses by Gender 

```python

def plot_donut_chart(df, category_col, title=None, hole=0.4, colors=None, width=600, height=550):
    """
    Creates a donut chart with percentage and absolute values displayed.

    Parameters:
    -----------
    df : pd.DataFrame
        Input dataframe
    category_col : str
        Column for which to plot counts as a donut chart.
    title : str, optional
        Chart title. Defaults to "Responses by {category_col}".
    hole : float, optional
        Fraction of the pie to cut out (0 < hole < 1). Default=0.4.
    colors : list, optional
        List of colors for categories. Default=px.colors.qualitative.Vivid.
    width, height : int
        Dimensions of the chart.

    Returns:
    --------
    fig : plotly.graph_objects.Figure
        Donut chart figure.
    """

    # Count occurrences and compute percentages
    counts = df[category_col].value_counts().reset_index()
    counts.columns = [category_col, 'Responses']
    total_responses = df.shape[0]
    counts['Percent'] = (counts['Responses'] / total_responses * 100).round(1)
    counts['Label'] = counts['Percent'].astype(str) + '% (' + counts['Responses'].astype(str) + ')'

    # Default colors if not provided
    if colors is None:
        colors = px.colors.qualitative.Vivid

    # Create donut chart
    fig = px.pie(
        counts,
        names=category_col,
        values='Responses',
        color=category_col,
        color_discrete_sequence=colors,
        hole=hole,
        title=title if title else f"Responses by {category_col}"
    )

    # Show labels with percentage + absolute value
    fig.update_traces(
        textinfo='label+percent+value',
        texttemplate='%{label}: %{percent} (%{value})'
    )

    # Layout adjustments
    fig.update_layout(
        width=width,
        height=height,
        margin=dict(t=80, l=50, r=150, b=50),
        title=dict(
            text=title if title else f"Responses by {category_col}",
            x=0.5,
            xanchor='center',
            font=dict(
                size=24,
                color='black',
                family="Arial Black, sans-serif"
            )
        )
    )

    return fig

# Creating the plot
fig_donut_gender = plot_donut_chart(df_clean, category_col="clean_gender", title="Responses by Gender")
fig_donut_gender.show()
```

#### 4- Responses by Age 

```python

def plot_age_group_rounded(df, age_col="Age", bins=[0,25,35,45,55,100],
                           labels=['<25','25-34','35-44','45-54','55+'],
                           title="Responses by Age Group", width=700, height=500):
    """
    Creates a cylinder-style column chart for age groups with counts and percentages.

    Parameters:
    -----------
    df : pd.DataFrame
        Input dataframe.
    age_col : str
        Column containing age values.
    bins : list
        Age bin edges.
    labels : list
        Labels for age groups corresponding to bins.
    title : str
        Chart title.
    width, height : int
        Chart dimensions.

    Returns:
    --------
    fig : go.Figure
        Rounded/cylinder-style column chart.
    """

    # Create Age Group column
    df['Age_Group'] = pd.cut(df[age_col], bins=bins, labels=labels, right=False)

    # Count responses by Age Group
    age_counts = (
        df['Age_Group']
        .value_counts()
        .sort_index()  # keep age groups in order
        .reset_index()
    )
    age_counts.columns = ['Age_Group', 'Count']

    # Compute percentages
    total_responses = df.shape[0]
    age_counts['Percent'] = (age_counts['Count'] / total_responses * 100).round(1)

    # Format labels: "xx% (count)"
    age_counts['Label'] = age_counts['Percent'].astype(str) + '% (' + age_counts['Count'].astype(str) + ')'

    # Build cylinder-style bar chart
    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=age_counts['Age_Group'],
        y=age_counts['Count'],
        text=age_counts['Label'],
        textposition='outside',
        marker=dict(
            color='teal',
            line=dict(color='black', width=1),
            cornerradius=[25,25,25,25]  # round all corners for cylinder effect
        ),
        width=0.5
    ))

    # Layout adjustments
    fig.update_layout(
        xaxis_title='Age Group',
        yaxis_title='Number of Responses',
        width=width,
        height=height,
        showlegend=False,
        title=dict(
            text=title,
            x=0.5,
            xanchor='center',
            font=dict(size=24, color='black', family="Arial Black, sans-serif")
        )
    )

    return fig

# Creating the plot
fig_age_bar = plot_age_group_rounded(df_clean)
fig_age_bar.show()
```

```python
# Function to plot cones types graphs

def plot_cones_graph(df, variable_col, treatment_col="treatment"):
    """
    Plots a 2D cone-style graph showing the impact of a variable on treatment.
    
    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe
    variable_col : str
        Column to compare against treatment (e.g., "benefits", "self_employed")
    treatment_col : str
        Treatment column name (default="treatment")
    
    Returns
    -------
    fig : go.Figure
        Plotly figure object with 2D cones
    """
    
    # Count treatment vs. variable
    counts = df.groupby([variable_col, treatment_col]).size().reset_index(name="count")
    
    # Calculate percentages (per variable group)
    counts["percent"] = counts.groupby(variable_col)["count"].transform(lambda x: 100 * x / x.sum())
    
    # Unique values
    categories = counts[variable_col].unique()
    treatments = counts[treatment_col].unique()
    colors = {"Yes": "teal", "No": "orange"}  # fallback for other treatment categories
    
    fig = go.Figure()
    spacing = 0.2  # horizontal spacing between cones
    
    # Draw cones
    for i, cat in enumerate(categories):
        for j, t in enumerate(treatments):
            val = counts[(counts[variable_col] == cat) & (counts[treatment_col] == t)]["percent"].values[0]
            x_center = i + j * spacing - spacing/2
            fig.add_trace(go.Scatter(
                x=[x_center - 0.2, x_center, x_center + 0.2, x_center - 0.2],
                y=[0, val, 0, 0],
                fill="toself",
                mode="lines+text",
                line=dict(color="black"),
                fillcolor=colors.get(t, "gray"),
                text=[None, f"{val:.1f}%", None, None],
                textposition="top center",
                name=t,
                showlegend=(i == 0)
            ))
    
    # Layout
    fig.update_layout(
        autosize=False,
        width=650,
        height=450,
        margin=dict(l=50, r=150, t=100, b=50),  # leave space for right legend
        title={
            "text": f"<b>Impact of {variable_col.capitalize()} on Treatment</b>",
            "x": 0.5,
            "xanchor": "center",
            "font": {"size": 22, "family": "Arial", "color": "black"}
        },
        xaxis=dict(
            title=variable_col.capitalize(),
            tickvals=list(range(len(categories))),
            ticktext=list(categories)
        ),
        yaxis=dict(title="Percentage of Responses", range=[0, 100]),
        legend=dict(
            orientation="v",      # vertical
            yanchor="top",
            y=1,                 # top of the graph
            xanchor="left",
            x=1.05               # slightly outside the right margin
        )
    )
    
    return fig
```

```python
# Function to plot cylindertypes graphs

def plot_rounded_bar(df, category_col, treatment_col="treatment"): 
    """
    Creates a stacked rounded bar chart with percentages for treatment vs. a chosen category.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Input dataframe
    category_col : str
        Column to compare against treatment (e.g., "benefits", "self_employed")
    treatment_col : str
        Treatment column name (default = "treatment")
    
    Returns:
    --------
    fig : go.Figure
        Plotly stacked bar chart
    """
    
    # Count treatment vs category
    counts = df.groupby([category_col, treatment_col]).size().reset_index(name="count")

    # Calculate percentages (per category group)
    counts["percent"] = counts.groupby(category_col)["count"].transform(lambda x: 100 * x / x.sum())

    # Build bar chart with rounded bars
    fig = go.Figure()

    for t in counts[treatment_col].unique():
        df_t = counts[counts[treatment_col] == t]
        fig.add_trace(go.Bar(
            x=df_t[category_col],
            y=df_t["percent"],
            name=t,
            text=df_t["percent"].round(1).astype(str) + "%",
            textposition="inside",
            marker=dict(
                color="teal" if t == "Yes" else "orange",
                line=dict(color="black", width=1),
                cornerradius="50%"
            ),
            width=0.5
        ))

    # Layout
    fig.update_layout(
        autosize=False,
        width=650,   
        height=450,
        margin=dict(l=50, r=150, t=100, b=50),  # leave space for legend
        barmode="stack",
        title={
            "text": f"<b>Impact of {category_col.capitalize()} on Treatment</b>",
            "x": 0.5,
            "xanchor": "center",
            "font": {"size": 22, "family": "Arial", "color": "black"}
        },
        xaxis_title=category_col.capitalize(),
        yaxis_title="Percentage of Responses",
        legend=dict(
            orientation="v",      # vertical
            yanchor="top",
            y=1,
            xanchor="left",
            x=1.05               # slightly outside the right margin
        )
    )
    
    return fig
```

### 4- Stigma & Perception


```python


def rounded_bar_plot_(df, category_col, seek_help_col="seek_help"):
    """
    Creates a stacked rounded bar chart with percentages for seek_help vs. a chosen category.
    Works for multiple values in seek_help (e.g., Yes, No, Maybe).

    Parameters:
    -----------
    df : pd.DataFrame
        Input dataframe
    category_col : str
        Column to compare against seek_help (e.g., "Anonymity", "self_employed")
    seek_help_col : str
        Column name for seek_help (default = "seek_help")

    Returns:
    --------
    fig : go.Figure
        Plotly stacked bar chart
    """
    
    # Count seek_help vs category
    counts = df.groupby([category_col, seek_help_col]).size().reset_index(name="count")

    # Calculate percentages (per category group)
    counts["percent"] = counts.groupby(category_col)["count"].transform(lambda x: 100 * x / x.sum())

    # Define colors for seek_help values
    base_colors = {"Yes": "teal", "No": "orange", "Maybe": "lightblue"}
    
    fig = go.Figure()

    for val in counts[seek_help_col].unique():
        df_val = counts[counts[seek_help_col] == val]
        fig.add_trace(go.Bar(
            x=df_val[category_col],
            y=df_val["percent"],
            name=val,
            text=df_val["percent"].round(1).astype(str) + "%",
            textposition="inside",
            marker=dict(
                color=base_colors.get(val, "steelblue"),  # fallback color if unknown value
                line=dict(color="black", width=1),
                cornerradius="50%"
            ),
            width=0.5
        ))

    # Layout with legend on the right
    fig.update_layout(
        autosize=False,
        width=650,
        height=450,
        margin=dict(l=50, r=150, t=100, b=50),  # extra right margin for legend
        barmode="stack",
        title={
            "text": f"<b>Impact of {category_col.capitalize()} on Seeking Help</b>",
            "x": 0.5,
            "xanchor": "center",
            "font": {"size": 22, "family": "Arial", "color": "black"}
        },
        xaxis_title=category_col.capitalize(),
        yaxis_title="Percentage of Responses",
        legend=dict(
            orientation="v",      # vertical legend
            yanchor="top",
            y=1,
            xanchor="left",
            x=1.05
        )
    )
    
    return fig
```

```python
import plotly.graph_objects as go

def plot_2d_cones(df, x_col, group_col, title=None, colors=None, spacing=0.4):
    """
    Plot 2D cone-like triangles for categorical data with percentages.

    Parameters:
        df : pd.DataFrame
            The dataframe containing the data.
        x_col : str
            The column for x-axis categories.
        group_col : str
            The column for grouping (different colors per value).
        title : str, optional
            The chart title.
        colors : dict, optional
            Dictionary mapping group values to colors.
        spacing : float
            Horizontal spacing between cones of different groups.
    """
    # Count occurrences
    counts = df.groupby([x_col, group_col]).size().reset_index(name="count")

    # Calculate percentages per x_col group
    counts["percent"] = counts.groupby(x_col)["count"].transform(lambda x: 100 * x / x.sum())

    # Unique values
    x_vals = counts[x_col].unique()
    groups = counts[group_col].unique()
    
    # Default colors if none provided
    if colors is None:
        default_colors = ["teal", "orange", "purple", "green", "blue"]
        colors = {g: default_colors[i % len(default_colors)] for i, g in enumerate(groups)}

    fig = go.Figure()

    # Draw cones
    for i, x_val in enumerate(x_vals):
        for j, g in enumerate(groups):
            val_series = counts[(counts[x_col] == x_val) & (counts[group_col] == g)]["percent"]
            if len(val_series) == 0:
                val = 0
            else:
                val = val_series.values[0]
            x_center = i + j*spacing - spacing/2
            fig.add_trace(go.Scatter(
                x=[x_center - 0.2, x_center, x_center + 0.2, x_center - 0.2],
                y=[0, val, 0, 0],
                fill="toself",
                mode="lines+text",
                line=dict(color="black"),
                fillcolor=colors[g],
                text=[None, f"{val:.1f}%", None, None],
                textposition="top center",
                name=g,
                showlegend=(i==0)
            ))

    # Layout
    fig.update_layout(
        title={"text": title or f"{group_col} by {x_col} (2D Cones)", "x":0.5, "xanchor":"center", "font":{"size":22}},
        xaxis=dict(title=x_col, tickvals=list(range(len(x_vals))), ticktext=list(x_vals)),
        yaxis=dict(title="Percentage of Responses", range=[0, 100]),
    )

    return fig
```
